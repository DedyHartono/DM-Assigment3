```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("covid19_indonesia1.csv")
data.head
```

```python
data.info()
```

```python
data['Date'].unique()
```

```python
latest_date = '9/15/2022'
data[data['Date']==latest_date]
```

```python
unique_countries = data[data.Latitude>0]['Location'].unique()
unique_countries.sort()
unique_countries
```

```python
country_Latitude_case = []
for i in unique_countries:
  country_Latitude_case.append(data[data.Latitude>0][data['Location']==i][data['Date']==latest_date].Latitude.sum())
```

```python
print(set(zip(unique_countries, country_Latitude_case)))
```

```python
unique_Location = data['Location'][data.Latitude>1].unique()
unique_Location
```

```python
Location_Latitude_case = []
for i in unique_Location:
  Location_Latitude_case.append(data[data.Latitude>0][data['Location']==i][data['Date']==latest_date].Latitude.sum())
```

```python
print(set(zip(unique_Location, Location_Latitude_case)))
```

```python
nan_indices = []
for i in range(len(unique_Location)):
  if type(unique_Location[i])==float:
    nan_indices.append
```

```python
unique_Location = list(unique_Location)
Location_Latitude_case = list(Location_Latitude_case)
```

```python
for i in nan_indices:
  unique_Location.pop(i)
  Location_Latitude_case.pop(i)

unique_Location
```

```python
plt.barh(unique_countries, country_Latitude_case)
plt.show()
```

```python
plt.figure(figsize=(50,40))
plt.barh(unique_Location, Location_Latitude_case)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(country_Latitude_case)
plt.legend(unique_countries, loc='best')
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.pie(Location_Latitude_case)
plt.legend(unique_Location, loc='best')
plt.show()
```
